```python
import pandas as pd
```

```python
#lendo os arquivos. Cada tabela esta em uma aba diferente no mesmo arquivo
df0= pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/vendas1.xlsx", sheet_name= 0)
df1= pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/vendas1.xlsx", sheet_name= 1)
df2= pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/vendas1.xlsx", sheet_name= 2)
df3= pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/vendas1.xlsx", sheet_name= 3)
#vamos juntar as tabelas em umas só
df= pd.concat([df0, df1, df2, df3])
df.sample(5)
```

```python
# Aqui temos a coluna data no formato correto, datetime. Mas vamos tratar a seguir como se não fosse e corrigi-lo
df.dtypes
```

```python
#transformando em datetime
df["Data"]= pd.to_datetime(df["Data"])
df.dtypes
```

```python
#Agrupando, descobrindo a reiceita anual

#criando a col
df["Receita"]= df["Qtde"]* df["Vendas"]

#agrupando
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#criando a coluna ANO
df["Ano_venda"]= df["Data"].dt.year
#criando coluna mes e dia
df["Mes_venda"], df["Dia_venda"]= (df["Data"].dt.month, df["Data"].dt.day)
df.sample(3)
```

```python
#É possivel usar algumas funções que usamos com numeros, como MIN, MAX 
print(df["Data"].min())
print(df["Data"].max())
#E realizar OPERAÇÃOES MATEMATICAS + - por exemplo
print(df["Data"].max()- df["Data"].min(), "\n")

#Para se referenciar a Data (ou qlqr outro dado) de forma automatica (como no excel)
  #basta criar uma coluna e atribuir como valor o nome da coluna e a operação que deseja realizar. EX:
df["Dif de dias"]= df["Data"]- df["Data"].min() #A data da venda - a menor data (a primeira venda)

#Usando a Função QUARTER para separa por trimestre
df["Trimestre_Venda"]= df["Data"].dt.quarter
df.sample(5)
```

```python
#FILTRAR as vendas de 2019 e do mes de março
vendas_marco_19= df.loc[(df["Data"].dt.year== 2022) & (df["Data"].dt.month== 5)]
vendas_marco_19.sample(5) #se o parametro da amostra for mais que a população dá erro
```

```python
dfVendas1 = pd.ExcelWriter('dfVendas1.xlsx')
df.to_excel(dfVendas1)
dfVendas1.save()
```
