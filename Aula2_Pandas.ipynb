```python
import pandas as pd
```

```python
#lendo os arquivos. Cada tabela esta em uma aba diferente no mesmo arquivo
df0= pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/vendas1.xlsx", sheet_name= 0)
df1= pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/vendas1.xlsx", sheet_name= 1)
df2= pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/vendas1.xlsx", sheet_name= 2)
df3= pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/vendas1.xlsx", sheet_name= 3)
```

```python
#vamos juntar as tabelas em umas só
df= pd.concat([df0, df1, df2, df3])
df.head()
```

```python
#pega uma amostra "aleatoria" da base
df.sample(6)
```

```python
#alterarando o tipo de dados das colunas
df["LojaID"]= df["LojaID"].astype("object")
#conferindo
df.dtypes
```

```python
#verificando se há valores nulos na tabela. E se houver mostrar a soma dessas ocorrencias
df.isnull().sum()
#Caso haja valroes nulos, usar a funcao "fillna" (substituir) ou "dropna"(exluir linha)
```

```python
#CRIANDO COLUNAS
#coluna receita= qnt vendas* valor unitario
df["Receita"]= df["Vendas"].mul(df["Qtde"])
df.head()
```

```python
maiorReceita= df["Receita"].max()
menorReceita= df["Receita"].min()
print(f"A maior receita foi R$ {maiorReceita} e a menor receita R${menorReceita}")
```

```python
#nlargest retorna os N maiores elementos de uma coluna
df.nlargest(4, "Receita")
```

```python
#nsmallest retorna os N menores elementos de uma coluna
df.nsmallest(4, "Receita")
```

```python
#agrupamento. Soma das vendas por estado
df.groupby("Cidade")["Receita"].sum()
```

```python
#ordenando o conjunto de dados com base em algum criterio
df.sort_values("Receita", ascending= False).head(10)
```
