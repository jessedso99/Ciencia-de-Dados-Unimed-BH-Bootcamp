```python
import pandas as pd
```

```python
df= pd.read_excel("/content/drive/MyDrive/Colab Notebooks/Datasets/dfVendas1.xlsx")
df.sample(5)
```

```python
#grafico de linhas (basta n informar qual tipo de plot vc quer. Apenas plot())
df.groupby('LojaID').agg(Qnt_Prod_Vendido=('Qtde', 'sum')).sort_values(by= ["Qnt_Prod_Vendido"], ascending= False).plot.bar()
df[["LojaID", "Qtde"]].loc[(df["LojaID"] == 1013)]
```

```python
#Grafico de Barras Vertical
df["LojaID"].value_counts(ascending= False).plot.bar()
```

```python
#Grafico de Barras Horizontal
df["LojaID"].value_counts(ascending= False).plot.barh()
```

```python
#Grafico de Pizza
df.groupby(df["Data"].dt.year)["Receita"].sum().plot.pie()
```

```python
#Para personalizar o garfico, mudando cores, titulos, etc, 칠 necessario importar a biblioteca matplotlib.pyplot
import matplotlib.pyplot as plt
```

```python

df["Cidade"].value_counts().plot.bar(title= "Total de vendas por Cidade", color= "red")
plt.xlabel("Estado")
plt.ylabel("Total Vendas")
plt.legend("Estados")
```

```python
#Podemos ainda alterar o estilo do grafico. Os apresentados anteriomente s칚o os Default.
#ver documenta칞a칫 do matplotlib
plt.style.use("ggplot")
df["Cidade"].value_counts().plot.barh(title= "Total de vendas por Cidade", color= "red")
plt.xlabel("Estado")
plt.ylabel("Total Vendas")
plt.legend();
```

```python
#quantidad de produtos vedidos por m칡s durante o ano de 2020
vendas2020= df[df["Ano_venda"]==2020]
vendas2020.groupby(vendas2020["Mes_venda"])["Qtde"].sum().plot(marker= "v")
plt.xlabel("M칡s")
plt.xlabel("Total Produtos vendidos")
plt.legend()
vendas2020.groupby(vendas2020["Mes_venda"])["Qtde"].sum()
```

```python
#Plotando HISTOGRAMA
plt.hist(df["Qtde"], color= "tan");
(df[["LojaID", "Qtde"]].loc[(df["Qtde"] <= df["Qtde"].max()) & (df["Qtde"] >= 30)]).shape
```

```python
#SCATTER PLOT, grafico de dispers칚o
plt.scatter(x= vendas2020["Dia_venda"], y= vendas2020["Receita"])
```

```python
#SALVANDO OS GRAFICOS usando plt.savefig("title.png")
vendas2020= df[df["Ano_venda"]==2020]
vendas2020.groupby(vendas2020["Mes_venda"])["Qtde"].sum().plot(marker= "v")
plt.xlabel("M칡s")
plt.xlabel("Total Produtos vendidos")
plt.legend()
vendas2020.groupby(vendas2020["Mes_venda"])["Qtde"].sum()
plt.savefig("QtdProd_vedido-mes_2020.png")
```
